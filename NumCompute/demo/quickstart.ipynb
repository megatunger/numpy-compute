## Loading data

`NumCompute` allows loading data with its `io` module. You can load data from any csv file using `load_csv`, carefully adjusting its argument to match your specific data source.

Below is an example where we load **Wine target dataset** from a csv file we provided, which contains 1599 samples with 12 real features. Since this file using semi-colon instead of commonly used colon, we have to specify delimiter as `";"`. The csv files are downloaded from [UCI Machine Learning Repository](https://archive.ics.uci.edu/dataset/186/wine+target)

```python
from numcompute.io import load_csv

wine = load_csv("../assets/winequality-red.csv", dtype=np.float32, delimiter=";")
wine.shape
```

## Pre-processing and Statistics

After loading data, you can exploring and processing your data using our `stats` and `preprocessing` modules.

```python
from numcompute.stats import mean, median, std, min, max
from numcompute.preprocessing import MinMaxScaler, StandardScaler
from numcompute.pipeline import Pipeline
```

Let's inspecting our data. You can compute the statistics (mean, median, std) of any array using `numcompute.stats` module as demonstrated below. Note how we specify the `axis=0` to correctly compute the statistics specific to each features.

```python
mean(wine, axis=0)
```

```python
median(wine, axis=0)
```

```python
std(wine, axis=0)
```

Let's also check the min and max values for each features.

```python
min(wine, axis=0), max(wine, axis=0)
```

Now, let's try pre-processing our data using `MinMaxScaler`, which scale each features into range [0,1].

```python
scaler = MinMaxScaler()
# scaler = StandardScaler()
wine = scaler.fit_transform(wine)
```

Let's re-check the min and max of our data after processing 

```python
min(wine, axis=0), max(wine, axis=0)
```

We can also chained processor into a Pipeline, which significantly simplify our code as we don't have to repeatedly call fit_transform.

```python
pipe = Pipeline([
    ("scaler1", StandardScaler()),
    ("scaler2", MinMaxScaler())
])

wine = pipe.fit_transform(wine)
```

```python
min(wine, axis=0), max(wine, axis=0)
```

## Sorting, Searching and Ranking

You can also sort and search any array with our `sort_search`, or rank the elements of an array with `rank`.

Let's try sorting our data with `multi_key_sort`, with which you can sort your your data using any features in any order. Note that here we are indexing `[:10]` to only show the first 10 records after sorted.

Or you can find out the numerical ranks to values of a feature.

Or you can compute the q-th percentile(s) of a feature

```python
percentile(wine[:, 0], q=98)
```

## Optimization

With `optim` module, you can easily compute gradients for iterative optimization algorithms.

```python
from numcompute.optim import grad
```

Let's try defining some random parameters and compute its gradient with respect to a simple MSE loss.

```python
from numcompute.metrics import mean_squared_error

# First we split our data into variables and target
variables = wine[:, :11]
target = wine[:, [11]]

# To compute the gradient of MSE loss with respecto parameters, we have to wrap
# every computation inside a single function
def forward(param):
    global variables, target
    batched_target = np.repeat(target, param.shape[0], axis=-1)
    pred = variables @ param.transpose()
    mse_loss = mean_squared_error(batched_target, pred, axis=0)
    return mse_loss

param = np.zeros(11)
g = grad(forward, param)
print("gradient", g)
```

Let's try updating our parameters using the computed gradients. Note how the updated parameters yield lower MSE loss, which demonstrate how this gradient can be used for optimization.

```python
print("initial param", param)
print("loss before", forward(param[np.newaxis]))
updated_param = param - 0.1 * g
print("updated param", updated_param)
print("loss after", forward(updated_param[np.newaxis]))
```

## Benchmark

Finally, let's run a benchmark to see how vectorization make everything faster. As can be seen in the displayed table, our module with vectorization can speed up operations by up to 300x, comparing to python loop. 

```python
from numcompute.benchmarking import main
from IPython.display import Markdown, display

results = main()

rows = []
for module, benchmarks in results.items():
    for benchmark_name, timings in benchmarks.items():
        rows.append(
            {
                "Module": module,
                "Benchmark": benchmark_name,
                "Loop (s)": timings["loop"],
                "Vectorized (s)": timings["vectorized"],
                "Speedup (x)": timings["speedup"],
            }
        )

rows.sort(key=lambda row: row["Speedup (x)"], reverse=True)

header = "| Module | Benchmark | Loop (s) | Vectorized (s) | Speedup (x) |"
separator = "|---|---|---:|---:|---:|"
body = [
    f"| {row['Module']} | {row['Benchmark']} | {row['Loop (s)']:.6f} | {row['Vectorized (s)']:.6f} | {row['Speedup (x)']:.2f} |"
    for row in rows
]

table_md = "\n".join([header, separator, *body])
display(Markdown(table_md))
```
