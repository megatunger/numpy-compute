## Loading data

`NumCompute` allows loading data with its `io` module. You can load data from any csv file using `load_csv`, carefully adjusting its argument to match your specific data source.

Below is an example where we load **Wine target dataset** from a csv file we provided, which contains 1599 samples with 12 real features. Since this file using semi-colon instead of commonly used colon, we have to specify delimiter as `";"`. The csv files are downloaded from [UCI Machine Learning Repository](https://archive.ics.uci.edu/dataset/186/wine+target)

```python
from numcompute.io import load_csv

wine = load_csv("../assets/winetarget-red.csv", dtype=np.float32, delimiter=";")
wine.shape
```

Let's take a quick look at our data. As you can see, our data is loaded as a 2d-array `wine` with `dtype=float32`

## Sorting, Searching and Ranking

You can also sort and search any array with our `sort_search`, or rank the elements of an array with `rank`.

```python
from numcompute.sort_search import multi_key_sort, topk
from numcompute.rank import rank, percentile
```

Let's try sorting our data with `multi_key_sort`, with which you can sort your your data using any features in any order. Note that here we are indexing `[:10]` to only show the first 10 records after sorted.

```python
multi_key_sort(wine, columns=[3,2,1])[:10]
```

To select the top-k values of any features, you can use `topk` function, which select the top-k values and also return the indices of the selected values.

```python
topk(wine[:, 0], k=5)
```

Or you can find out the numerical ranks to values of a feature.

```python
rank(wine[:, 0], method="ordinal")
```

Or you can compute the q-th percentile(s) of a feature

```python
percentile(wine[:, 0], q=98)
```

## Optimization

With `optim` module, you can easily compute gradients for iterative optimization algorithms.

```python
from numcompute.optim import grad
```

Let's try defining some random parameters and compute its gradient with respect to a simple MSE loss.

```python
from numcompute.metrics import mean_squared_error

# First we split our data into variables and target
variables = wine[:, :11]
target = wine[:, [11]]

# To compute the gradient of MSE loss with respecto parameters, we have to wrap
# every computation inside a single function
def forward(param):
    global variables, target
    batched_target = np.repeat(target, param.shape[0], axis=-1)
    pred = variables @ param.transpose()
    mse_loss = mean_squared_error(batched_target, pred, axis=0)
    return mse_loss

param = np.random.rand(11)
g = grad(forward, param)
print("gradient", g)
```

Let's try updating our parameters using the computed gradients. Note how the updated parameters yield lower MSE loss, which demonstrate how this gradient can be used for optimization.

```python
print("initial param", param)
print("loss before", forward(param[np.newaxis]))
updated_param = param - 0.1 * g
print("updated param", updated_param)
print("loss after", forward(updated_param[np.newaxis]))
```
