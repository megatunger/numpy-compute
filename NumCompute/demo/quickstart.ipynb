## Loading data

`NumCompute` allows loading data with its `io` module. You can load data from any csv file using `load_csv`, carefully adjusting its argument to match your specific data source.

Below is an example where we load **Wine Quality dataset** from a csv file we provided, which contains 1599 samples with 12 real features. Since this file using semi-colon instead of commonly used colon, we have to specify delimiter as `";"`. The csv files are downloaded from [UCI Machine Learning Repository](https://archive.ics.uci.edu/dataset/186/wine+quality)

```python
from numcompute.io import load_csv

wine = load_csv("../assets/winequality-red.csv", dtype=np.float32, delimiter=";")
wine.shape
```

Let's take a quick look at our data. As you can see, our data is loaded as a 2d-array `wine` with `dtype=float32`

```python
wine
```

## Pre-processing and Statistics

After loading data, you can exploring and processing your data using our `stats` and `preprocessing` modules.

```python
from numcompute.stats import mean, median, std, min, max
from numcompute.preprocessing import MinMaxScaler
```

Let's inspecting our data. You can compute the statistics (mean, median, std) of any array using `numcompute.stats` module as demonstrated below. Note how we specify the `axis=1` to correctly compute the statistics specific to each features.

```python
mean(wine, axis=1)
```

```python
median(wine, axis=1)
```

```python
std(wine, axis=1)
```

Let's also check the min and max values for each features.

```python
min(wine, axis=1), max(wine, axis=1)
```

Now, let's try pre-processing our data using `MinMaxScaler`, which scale each features into range [0,1].

```python
scaler = MinMaxScaler()
wine = scaler.fit_transform(wine)
```

Let's re-check the min and max of our data after processing 

```python
min(wine, axis=1), max(wine, axis=1)
```

## Sorting, Searching and Ranking

You can also sort and search any array with our `sort_search`, or rank the elements of an array with `rank`.

```python
from numcompute.sort_search import multi_key_sort, topk
from numcompute.rank import rank, percentile
```

Let's try sorting our data with `multi_key_sort`, with which you can sort your your data using any features in any order. Note that here we are indexing `[:10]` to only show the first 10 records after sorted.

```python
multi_key_sort(wine, columns=[3,2,1])[:10]
```

To select the top-k values of any features, you can use `topk` function, which select the top-k values and also return the indices of the selected values.

```python
topk(wine[:, 0], k=5)
```

Or you can find out the numerical ranks to values of a feature.

```python
rank(wine[:, 0], method="ordinal")
```

Or you can compute the q-th percentile(s) of a feature

```python
percentile(wine[:, 0], q=98)
```
